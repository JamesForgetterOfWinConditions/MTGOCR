```python
from PIL import Image
from fuzzywuzzy import fuzz
import re
import pytesseract

pytesseract.pytesseract.tesseract_cmd = "C:\\Program Files\\Tesseract-OCR\\tesseract.exe"
```

```python
def remove_non_alphabetic(text):
    return re.sub("[^a-zA-Z]", " ", text)
```

```python
def filter_names(tokens, min_length=2, max_length=15):
    return [token for token in tokens if min_length <= len(token) <= max_length]
```

```python
file_path = r"C:\Users\james\Downloads\PostpunkRevivalPoweredCube.txt"  # Replace 'your_file.txt' with the actual file path

# Setting the list of Viable names
with open(file_path, 'r') as file:
    # Use readlines() to read all lines in the file and store them in a list
    name_pool = file.readlines()
# name_pool = ["Vampire Nighthawk", "Ardenvale Tactician", "Dauthi Mindripper", "Ravenous Chupacabra", "Gravedigger"]

# Set a threshold for matching (adjust as needed)
threshold = 1

# Function to find the best match
def find_best_match(extracted_name, name_pool):
    best_match = None
    max_similarity = 0

    for name in name_pool:
# Experimenting with best similarity matching tool.  fuzz.ratio, fuzz.partial_ratio, fuzz.token_sort_ratio
        similarity = fuzz.token_sort_ratio(extracted_name, name)
        
        if similarity > max_similarity and similarity >= threshold:
            max_similarity = similarity
            best_match = name

    return best_match, max_similarity
```

```python
ocr_output = pytesseract.image_to_string(Image.open(r"C:\Users\james\Downloads\test5.jpg"), lang='eng', config='--psm 12', nice=0, timeout=0)
print(ocr_output)
```

```python
# Name Processing
# ocr_output = "Name: Charizard 123"
cleaned_text = remove_non_alphabetic(ocr_output)
tokens = cleaned_text.split()
filtered_names = filter_names(tokens)
best_match, similarity = find_best_match(" ".join(filtered_names), name_pool)

if best_match is not None:
    print(f"Match found: {best_match} (Similarity: {similarity}%).")
else:
    print("No match found.")
```

```python
def find_best_match(extracted_name, name_pool):
    best_match = None
    max_similarity = 0

    for name in name_pool:
        similarity = fuzz.partial_ratio(extracted_name, name)
        
        if similarity > max_similarity:
            max_similarity = similarity
            best_match = name

    return best_match, max_similarity
```

```python
def find_best_match_with_sort(extracted_name, name_pool):
    best_match = None
    max_similarity = 0

    for name in name_pool:
        similarity = fuzz.token_sort_ratio(extracted_name, name)
        
        if similarity > max_similarity:
            max_similarity = similarity
            best_match = name

    return best_match, max_similarity
```
